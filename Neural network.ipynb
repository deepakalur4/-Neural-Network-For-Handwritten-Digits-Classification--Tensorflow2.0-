```python
import tensorflow
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
(x_train,y_train),(x_test,y_test)=tensorflow.keras.datasets.mnist.load_data()
```

```python
for i in range(6):
    print(y_train[i])
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train_flattned=x_train.reshape(len(x_train),28*28)
x_test_flattned=x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattned.shape
```

```python
x_test_flattned.shape
```

```python
x_train_flattned[0]
```

```python
from tensorflow import keras
```

```python
model=keras.models.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation="sigmoid")

])

model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model.fit(x_train_flattned,y_train,epochs=5)
```

```python

model.evaluate(x_test_flattned,y_test)
```

```python
y_pred=model.predict(x_test_flattned)
```

```python
y_pred[0]
```

```python
plt.matshow(x_test[0])
```

```python
np.argmax(y_pred[0])
```

```python
plt.matshow(x_test[1])
```

```python
np.argmax(y_pred[1])
```

```python
y_prediction=[np.argmax(i) for i in y_pred]
```

```python
tensorflow.math.confusion_matrix(labels=y_test,predictions=y_prediction)
```

```python
confusion_matricx=tensorflow.math.confusion_matrix(labels=y_test,predictions=y_prediction)
```

```python
sns.heatmap(confusion_matricx,annot=True,fmt="d",cmap="viridis")
plt.xlabel("predicted")
plt.ylabel("truth")
```

```python
model=keras.models.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation="relu"),
    keras.layers.Dense(10,activation="sigmoid")

])

model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model.fit(x_train_flattned,y_train,epochs=5)
```

```python

model.evaluate(x_test_flattned,y_test)
```

```python
y_pred=model.predict(x_test_flattned)
```

```python
y_prediction=[np.argmax(i) for i in y_pred]
```

```python
confusion_matricx=tensorflow.math.confusion_matrix(labels=y_test,predictions=y_prediction)
```

```python
sns.heatmap(confusion_matricx,annot=True,fmt="d",cmap="viridis")
plt.xlabel("predicted")
plt.ylabel("truth")
```
